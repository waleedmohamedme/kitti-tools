```python
from kitti import *
import cv2 
import matplotlib.pyplot as plt
```

```python
dataset=kitti()
dataset.base_dir='/root/hard/work/data/kitti/raw/mini/'
dataset.build()
```

```python
test_filepath = dataset.list_of_velo_files[66]
```

```python
def draw_images_with_annotations(max_to_draw=10):
    for i,test_filepath in enumerate(dataset.list_of_velo_files[:max_to_draw]):
        plt.figure(figsize=(15,10))
        test_scanpoint=dataset.load_scan_point_from_file(test_filepath)
        image=dataset.velo_to_image(test_scanpoint)
        velo_xmin=dataset.annotation[i][0]['proj_velo_xmin']
        velo_xmax=dataset.annotation[i][0]['proj_velo_xmax']
        velo_ymin=dataset.annotation[i][0]['proj_velo_ymin']
        velo_ymax=dataset.annotation[i][0]['proj_velo_ymax']
        cv2.rectangle(image,(velo_xmin,velo_ymin),(velo_xmax,velo_ymax),(1,0,1),2)
        plt.imshow(image)
        plt.show()
```

```python
draw_images_with_annotations()
```

```python
image = cv2.imread(dataset.save_dir+'testscanpoint.png')
cv2.rectangle(image,(velo_xmin,velo_ymin),(velo_xmax,velo_ymax),(255,255,0),2)
plt.imshow(image)
plt.show()
```
